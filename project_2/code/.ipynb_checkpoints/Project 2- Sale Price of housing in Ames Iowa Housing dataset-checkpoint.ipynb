# Problem Statement

This project seeks to identify the best regression model to predict the price of the housing in Ames, Iowa and the best predictors of housing prices. This can be done by looking into the house characteristics and locations.

Some of the predicted variables that can play an important role in this problem is OverallQual, YearBuilt, GrLivArea pertaining to the infrastructural and space aspect of the house. Location might be an important predictor and this will be investigated upon too. 

Real estate agencies with a separate database on house characteristics and locations will be able to estimate what price one of these houses will most likely sell for based on the model that is built using this data.

In performing Multiple Linear Regression, some underlying assumptions are made. One of which is the independence of predictors. There appears to be multicollinearity.

When there is multicollinearity, changes in one variable would also cause shifts in another variable. The stronger this correlation is, higher would be the difficulty to change one variable without changing another. This makes it difficult for the regression model to estimate the relationship between each independent variable and the dependent variable independently.

3 major problems will arise, which are related to each other:
1. Unreliable explained variance due to predictors
2. The coefficients become highly sensitive to small changes in the model and the presence of other independent predictors.
3. Reduction in the precision of the equation coefficients, thus reducing the performance of the regression model.

It would be useful to consider that GarageCars and GarageArea are highly correlated to each other since the number of cars that will fit into the garage will depend on GarageArea. This shows a clear case of multicollinearity. 

Also, TotalBsmtSF and 1stFlrSF show high correlation again indicating multicollinearity. 

Another set of variables that show high correlation are YearBuilt and GarageYrBlt. Considering that GarageYrBlt has over 5% missing values and YearBuilt has a relatively higher correlation score with SalePrice, GarageYrBlt should be removed

To further understand the correlation between categorical variables and SalePrice, the ANOVA test will be used. This test is used to determine if there exists significant difference in the mean of groups.

$H_0$ : There exists no significant difference between the groups

$H_A$ : There exists a significant difference between the groups

A significance level of $\alpha = 0.05$ will be used. If $p\text{-value} < \alpha$, then there is evidence to reject the null hypothesis, $H_0$ is accepted as incorrect and therefore $H_A$ is correct.


```python
cat_data = train_df.select_dtypes(exclude=[np.number])
cat_list = [col for col in cat_data.columns]
cat_list.extend(['SalePrice'])
cat_list
cat_data_new = train_df[cat_list].copy()

def anova(frame):
    anv = pd.DataFrame()
    anv['feature'] = cat_list
    pvals = []
    for c in cat_list:
        samples = []
        for cls in frame[c].unique():
            s = frame[frame[c] == cls]['SalePrice'].values
            samples.append(s)
        pval = stats.f_oneway(*samples)[1]
        pvals.append(pval)
    anv['pval'] = pvals
    return anv.sort_values('pval')

a = anova(cat_data_new)
a

#a['disparity'] = np.log(1./a['pval'].values) 
#sns.barplot(data=a, x = 'feature', y='disparity') 
#plt.xticks(rotation=90) 
#plt 
```

Using the ANOVA test, it can be seen that among all the categorical variables, Neighborhood is a significant feature followed by ExterQual, KitchenQual, etc.

Categorical Predictors that have more than 95% of their values in one category will be excluded: Street, Land Slope, Condition 2, Roof Matl, Heating. These predictors will not be included in modelling.

## Lot Frontage

Out of 2051 houses 330 of them had no value in lot frontage. One of the theory that could explain it is that all the null plots are high-rise buildings. Further investigation into the MS Subclass column of the dataset shows that none of them are multi-residential and 131 of them are marked as 1-STORY 1946 & NEWER ALL STYLES (131 ) so they can not all be condos.

As the theories are not evident in this dataset, an assumption that the values are missing completely at random. And as such the approach adopted is to create a model to predict the lot frontage from features that are likely to influence lot frontage

Low lot frontage appears to correspond with low lot area. There is also heteroscedacity in this relationship.The variance of Lot Frontage is not uniform accross the range of LotArea. The larger the overall area of the lot, the more variation there is in LotFrontage.

Townhouses are all clustered at the low end of both lot area and frontage. Duplex and single family homes have a much wider range of LotFrontage values.

Based on neighbourhood spreads of lot frontage, there are some areas in the city of Ames that have a fairly distinct spread of LotFrontage values. 

Properties with no access to Alley tend to have a higher median LotFrontage.

The median value of LotFrontage increases with every additional car in the garage, up to 3. 

The missing lot frontage values will be imputed subsequently by using a lasso model taking into consideration the predictors above.

```python
# LotArea vs. LotFrontage
def idOutliers (dat):
    tile25 = dat.describe()[4]
    tile75 = dat.describe()[6]
    iqr = tile75 - tile25 
    out = (dat > tile75+1.5*iqr) | (dat < tile25-1.5*iqr)
    return out
fig, ax =plt.subplots(1,2, figsize=(16,4))
ax[0].set_title('With LotArea outliers')
ax[1].set_title('Without LotArea outliers')
sns.regplot(train_df['Lot Area'], train_df['Lot Frontage'], ax=ax[0])
ax[0].set(xlabel='LotArea')
sns.regplot(train_df['Lot Area'][~idOutliers(train_df['Lot Area'])], train_df['Lot Frontage'][~idOutliers(train_df['Lot Area'])], ax=ax[1])
ax[1].set(xlabel='LotArea');
```

## Model fit and evaluation

To calculate the error for evaluation, a function get_metrics will be created.

```python
#Validation function
import sklearn.metrics as metrics
def r2_adj(y_true, y_preds, p):
    n = len(y_true)
    y_mean = np.mean(y_true)
    numerator = np.sum(np.square(y_true - y_preds)) / (n - p - 1)
    denominator = np.sum(np.square(y_true - y_mean)) / (n - 1)
    return 1 - numerator / denominator
def get_metrics(y_true, y_pred, p):
    mse = metrics.mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    msle = metrics.mean_squared_log_error(y_true, y_pred)
    mae = metrics.median_absolute_error(y_true, y_pred)
    r2 = metrics.r2_score(y_true, y_pred)
    r2a = r2_adj(y_true, y_pred, p)
    
    print('Mean squared error      = ', mse)
    print('Root mean squared error = ', rmse)
    print('Mean squared log error  = ', msle)
    print('Median absolute error   = ', mae)
    print('R^2                     = ', r2)
    print('Adjusted R^2            = ', r2a)
```
